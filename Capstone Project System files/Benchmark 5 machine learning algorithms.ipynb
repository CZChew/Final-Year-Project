# Sklearn to make the prediction model

```python
# Scaling the data between 0 and 1.
from sklearn import preprocessing
minmax = preprocessing.MinMaxScaler(feature_range=(0,1))
minmax.fit(x).transform(x)
```

```python
# separating the dataset into training dataset and test dataset. 
from sklearn import neighbors
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state = 0)
x_train.head()
```

```python
x_train.dtypes
```

```python
x_test.head()
```

```python
# type casting.
y_train=y_train.astype('int')
y_train
```

```python
y_test=y_test.astype('int')
y_test
```

```python
print("No. of null values: ", data.isnull().sum())
print("Filling null values with mean of that column:")
data = data.fillna(np.mean(data))
data.info()
```

# Machine Learning Algorithms

# K-Nearest Neighbour

```python
from sklearn.neighbors import KNeighborsClassifier
clf=neighbors.KNeighborsClassifier()
knn_model = clf.fit(x_train,y_train)
```

```python
# Predicted chance of Flood.
print("Predicted Values for the Floods:")
y_predict=clf.predict(x_test)
y_predict
```

```python
print("Actual Values:")
print(y_test.values)
```

```python
print("List of the Predicted Values:")
print(y_predict)
```

```python
# Scaling the dataset.
x_train_std= minmax.fit_transform(x_train)
x_test_std= minmax.fit_transform(x_test)
```

```python
# the accuracy of the model
from sklearn.metrics import accuracy_score
y_predict_knn = knn_model.predict(x_test)
print("\nAccuracy: %f"%(accuracy_score(y_test,y_predict_knn)*100))
```

```python
from sklearn.metrics import r2_score, mean_squared_error
r2scr = r2_score(y_test,y_predict_knn)
print("R2_score = ", r2scr)
mse = mean_squared_error(y_test,y_predict_knn)
print("\nMean squared error = ", mse)
rmse = np.sqrt(mean_squared_error(y_test,y_predict_knn))
print("\nRoot mean squared error = ", rmse)
```

# Support Vectors Classification

```python
from sklearn.svm import SVC
svc=SVC(kernel='rbf',probability=True)
svc_classifier=svc.fit(x_train,y_train)
```

```python
y_predict_svc=svc_classifier.predict(x_test)
print("Actual Flood Values: ", y_test.values)
print("Predicted Flood Values: ", y_predict_svc)
```

```python
# the accuracy of the model
from sklearn.metrics import accuracy_score
print("\nAccuracy: %f"%(accuracy_score(y_test,y_predict_svc)*100))
```

```python
from sklearn.metrics import r2_score, mean_squared_error
r2scr = r2_score(y_test,y_predict_svc)
print("R2_score = ", r2scr)
mse = mean_squared_error(y_test,y_predict_svc)
print("\nMean squared error = ", mse)
rmse = np.sqrt(mean_squared_error(y_test,y_predict_svc))
print("\nRoot mean squared error = ", rmse)
```

# Logistic Regression

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(x_train_std,y_train)
log_acc = cross_val_score(log,x_train_std,y_train,cv=10,scoring='accuracy',n_jobs=1)
log_acc
```

```python
y_predict_log = log.predict(x_test)
print("Actual Flood Values: ", y_test.values)
print("Predicted Flood Values: ", y_predict_log)
```

```python
# the accuracy of the model
from sklearn.metrics import accuracy_score
print("\nAccuracy score: %f"%(accuracy_score(y_test,y_predict_log)*100))
```

```python
from sklearn.metrics import r2_score, mean_squared_error
r2scr = r2_score(y_test,y_predict_log)
print("R2_score = ", r2scr)
mse = mean_squared_error(y_test,y_predict_log)
print("\nMean squared error = ", mse)
rmse = np.sqrt(mean_squared_error(y_test,y_predict_log))
print("\nRoot mean squared error = ", rmse)
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dt_clf=DecisionTreeClassifier()
dt_clf.fit(x_train,y_train)
dt_clf_acc=cross_val_score(dt_clf,x_train_std,y_train,cv=3,scoring="accuracy",n_jobs=-1)
dt_clf_acc
```

```python
print("Predicted Values:")
y_prediction_dt=dt_clf.predict(x_test)
y_prediction_dt
```

```python
print("Actual Values:")
print(y_test.values)
```
